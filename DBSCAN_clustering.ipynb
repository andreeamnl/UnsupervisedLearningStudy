# DBSCAN Clustering

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## Importing the dataset

```
dataset = pd.read_csv("Mall_Customers.csv")
X = dataset.iloc[:, [3,4]].values
```

## Feature Scaling

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X)
```

```
X_train
```

# Run DBSCAN

```
from sklearn.cluster import DBSCAN

db = DBSCAN(eps = 0.8, min_samples = 3)
db.fit(X_train)
labels = db.labels_
labels
```

```
df = pd.DataFrame(dataset, columns=["Annual Income (k$)","Spending Score (1-100)"])

df["cluster"] = labels
df["cluster"].value_counts()
```

```
df
```

```
unique_labels = np.unique(df["cluster"])
colors = plt.cm.viridis(np.linspace(0, 1, len(unique_labels)))

for lbl, col in zip(unique_labels, colors):
    mask = df["cluster"] == lbl
    label_name = f"Cluster {lbl}" if lbl != -1 else "Noise"
    plt.scatter(
        df.loc[mask, "Annual Income (k$)"].astype(float),
        df.loc[mask, "Spending Score (1-100)"].astype(float),
        c=[col],
        label=label_name,
        edgecolor='k',
        s=100
    )

plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.title("DBSCAN Clusters")
plt.legend()
plt.show()
```

## Using the elbow method to find the optimal number of clusters

```
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt
import numpy as np

min_samples = 3
neighbors = NearestNeighbors(n_neighbors=min_samples)
neighbors_fit = neighbors.fit(X_train)
distances, _ = neighbors_fit.kneighbors(X_train)

k_distances = np.sort(distances[:, min_samples-1])
plt.plot(k_distances)
plt.ylabel(f"{min_samples}-NN distance")
plt.xlabel("Points sorted by distance")
plt.title("k-distance Graph for DBSCAN")
plt.show()
```

## Training the DBSCAN model on the dataset

```
db = DBSCAN(eps=0.35, min_samples=3)
labels = db.fit_predict(X_train)

print("Cluster labels:", labels)
```

```
df = pd.DataFrame(dataset, columns=["Genre","Age","Annual Income (k$)","Spending Score (1-100)"])

df["cluster"] = labels
df["cluster"].value_counts()

```

```
df
```

```
unique_labels = np.unique(labels)
colors = plt.cm.viridis(np.linspace(0, 1, len(unique_labels)))

for lbl, col in zip(unique_labels, colors):
    mask = df["cluster"] == lbl
    label_name = f"Cluster {lbl}" if lbl != -1 else "Noise"
    plt.scatter(
        df.loc[mask, "Annual Income (k$)"].astype(float),
        df.loc[mask, "Spending Score (1-100)"].astype(float),
        c=[col],
        label=label_name,
        edgecolor='k',
        s=100
    )

plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.title("DBSCAN Clusters")
plt.legend()
plt.show()
```

```
from sklearn.metrics import silhouette_score

score = silhouette_score(X, labels)
print("Silhouette Score:", score)
```
